```
ab.sort_values(by="reviews_per_month", ascending=False, inplace=True)
ab.head(20)
```

```
ab.sort_values(by="number_of_reviews", ascending=False, inplace=True)
ab.head(20)
```

```
ab.sort_values(by="price", ascending=False, inplace=True)
ab.head(20)
```

```
print("number of different cities :", len(ab['city'].value_counts()))
print("number of different neighbourhood group :", len(ab['neighbourhood_group'].value_counts()))
print("number of different room type :", len(ab['room_type'].value_counts()))
```

```
ab['city'].value_counts().plot.bar(figsize=(10,5))
plt.title('Number of Airbnb per city')
plt.xlabel('City')
plt.ylabel('Number of Airbnb')
```

```
ab['room_type'].value_counts().plot.bar(figsize=(10,5))
plt.title('Number of Airbnb per room type')
plt.xlabel('room type')
plt.ylabel('Number')
```

```
ab[['room_type', 'price']].groupby('room_type').mean().plot.bar(figsize=(10,5))
plt.title('mean price of Airbnb per room type')
plt.xlabel('room type')
plt.ylabel('price')
```

```
ab[['city','reviews_per_month']].groupby('city').agg(nb_Airbnb=('city', 'count'), nb_review_per_month=('reviews_per_month', 'count')).sort_values(by="nb_Airbnb", ascending=False, inplace=False).plot.bar(figsize=(10,5))
plt.title('number of Airbnb and review per month')
plt.xlabel('City')
plt.ylabel('number of Airbnb and number of review per month')
```

```
ab['last_review'] = pd.to_datetime(ab['last_review'])
day_of_data_collection = ab['last_review'].max()
day_of_data_collection
```

```
pd.get_dummies(ab[['city', 'room_type']], columns=['room_type']).groupby('city').sum().sort_values(by='room_type_Entire home/apt', ascending=False, inplace=False).plot.bar(figsize=(20,5))
plt.title('room type of Airbnb per city per room type')
plt.xlabel('City')
plt.ylabel('number of room')
```

```
ab[['city', 'availability_365', 'minimum_nights']].groupby('city').mean().sort_values(by='availability_365', ascending=False, inplace=False).plot.bar(figsize=(17,5))
plt.title('mean availability and minimum nights per city')
plt.xlabel('City')
plt.ylabel('number of day')
```
